```
# Medical Sales Analysis Project
# Author: Tejas Patil
# Description: Analyze medical sales data to generate business insights

import pandas as pd
import matplotlib.pyplot as plt


def load_data(file_path: str) -> pd.DataFrame:
    """Load medical sales data from CSV file."""
    try:
        return pd.read_csv(file_path)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return pd.DataFrame()


def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """Clean and preprocess the dataset."""
    df = df.dropna()
    df["Order_Date"] = pd.to_datetime(df["Order_Date"])
    df["Sales"] = df["Sales"].astype(float)
    return df


def analyze_sales(df: pd.DataFrame) -> None:
    """Perform sales analysis."""
    print("Total Sales:", df["Sales"].sum())
    print("Top Selling Product:")
    print(df.groupby("Product")["Sales"].sum().idxmax())
    print("\nSales by Category:")
    print(df.groupby("Category")["Sales"].sum())


def visualize_sales(df: pd.DataFrame) -> None:
    """Visualize monthly sales trend."""
    monthly_sales = df.resample("M", on="Order_Date")["Sales"].sum()
    monthly_sales.plot()
    plt.title("Monthly Medical Sales Trend")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.show()


def main():
    # âœ… FILE PATH
    file_path = "medical_sales.csv"

    data = load_data(file_path)
    if data.empty:
        return

    data = clean_data(data)
    analyze_sales(data)
    visualize_sales(data)


if __name__ == "__main__":
    main()
```

```
def monthly_sales_trend(df):
    monthly_sales = df.resample("M", on="Order_Date")["Sales"].sum()
    monthly_sales.plot()
    plt.title("Monthly Medical Sales Trend")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.show()

monthly_sales_trend(data)
```

```
file_path = "medical_sales.csv"
data = load_data(file_path)
if not data.empty:
    data = clean_data(data)

def sales_by_product(df):
    product_sales = df.groupby("Product")["Sales"].sum()
    plt.figure()
    product_sales.plot(kind="bar")
    plt.title("Sales by Product")
    plt.xlabel("Product")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


sales_by_product(data)
```

```
def sales_by_region(df):
    region_sales = df.groupby("Region")["Sales"].sum()
    region_sales.plot(kind="bar")
    plt.title("Region-wise Medical Sales")
    plt.xlabel("Region")
    plt.ylabel("Total Sales")
    plt.show()

sales_by_region(data)
```

```
def sales_by_category(df):
    category_sales = df.groupby("Category")["Sales"].sum()
    category_sales.plot(kind="pie", autopct="%1.1f%%")
    plt.title("Sales Distribution by Category")
    plt.ylabel("")
    plt.show()

sales_by_category(data)
```
